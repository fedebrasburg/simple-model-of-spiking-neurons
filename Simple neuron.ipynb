```python
from sympy import Function, dsolve, Eq, Derivative, sin, cos, symbols, Symbol, evalf

t = Symbol('t')
v = Function('v')
u = Function('u')

I = 10
a = 0.02
b = 0.2
c = -65
d = 2

#Expresamos la ecuacion 
eq1 = Eq(Derivative(v(t), t), 0.04*(v(t)**2) + 5*v(t) + 140 - u(t) + I)
eq2 = Eq(Derivative(u(t), t), a*((b*v(t)) - u(t)))

f = dsolve((eq1, eq2), ics={v(0):-63, u(0):d})
v, u = f
v
```

```python
import matplotlib.pyplot as plt
```

```python
# 0.1ms
h = 0.1

def nextPoint(v, u, I):
    newV = v + h * (0.04*v*v + 5*v + 140 - u + I)
    newU = u + h * (a*(b*v - u))
    return newV,newU

def graph(a, b, c, d):
    v = [-65]
    u = [0]
    time = [0]
    n = 5000
    for i in range(n):
        time.append(i*h)
        if(i > n*0.25):
            I = 10
        else:
            I = 0
        newV, newU = nextPoint(v[i], u[i], I)
        if newV > 30:
            v.append(c)
            u.append(newU +d)
        else:
            v.append(newV)
            u.append(newU)

    plt.subplots(figsize=(14,4))
    plt.plot(time, v)
    plt.show()
```

```python
a = 0.02
b = 0.2
c = -65
d = 8
# Regular spiking (RS)
graph(a, b, c , d)
```
