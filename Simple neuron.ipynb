```python
import matplotlib.pyplot as plt
```

```python
# 0.1ms
h = 0.1

def nextPoint(v, u, I):
    newV = v + h * (0.04*v*v + 5*v + 140 - u + I)
    newU = u + h * (a*(b*v - u))
    return newV,newU

def graph(a, b, c, d):
    v = [-65]
    u = [v[0]*b]
    time = [0]
    n = 5000
    for i in range(n):
        time.append(i*h)
        if(i > n*0.25):
            I = 10
        else:
            I = 0
        newV, newU = nextPoint(v[i], u[i], I)
        if newV > 30:
            v.append(c)
            u.append(newU +d)
        else:
            v.append(newV)
            u.append(newU)

    plt.subplots(figsize=(14,4))
    plt.plot(time, v)
    plt.show()
```

```python
a = 0.02
b = 0.2
c = -65
d = 8
# Regular spiking (RS)
graph(a, b, c , d)
```
