```python
import matplotlib.pyplot as plt

# 0.1ms
h = 0.1

def nextPoint(v, u, I):
    newV = v + h * (0.04*v*v + 5*v + 140 - u + I)
    newU = u + h * (a*(b*v - u))
    return newV,newU

def graph(a, b, c, d, intensidadMaxima = 10, intensidadMinima = 0):
    v = [c]
    u = [v[0]*b]
    intensidad = [intensidadMinima]
    time = [0]
    n = 5000
    for i in range(n):
        time.append(i*h)
        if(i > n*0.25):
            I = intensidadMaxima
        else:
            I = intensidadMinima
        # Para el RZ
#         if(i > n*0.5 and i < n*0.55):
#             I *= 1.5
        intensidad.append(I)
        newV, newU = nextPoint(v[i], u[i], I)
        if newV > 30:
            v.append(c)
            u.append(newU +d)
        else:
            v.append(newV)
            u.append(newU)
    
    fig, (graficoV, graficoI) = plt.subplots(2, figsize=(14,8), ncols = 1,  gridspec_kw={'height_ratios':[4, 2]})
#     gs = gridspec.GridSpec(5,1)

    graficoV.plot(time, v)
    graficoV.set_title("V(t)")
    
    graficoI.plot(time, intensidad, 'tab:orange')
    graficoI.set_ylim(-2, intensidadMaxima + 10)
    graficoI.set_title("I(t)")
    
    plt.show()
```

```python
a = 0.02
b = 0.2
c = -65
d = 8

I = 10
# Regular spiking (RS)
graph(a, b, c , d, I)
```
